```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("posts.csv")
```

```python
scores = np.array(df['score'])
fig, ax = plt.subplots(1, 1)
ax.hist(scores)
ax.set_xlabel('Score')
ax.set_ylabel('Number of posts')

plt.show()
```

```python
data = np.array([])
for i in range(100):
    post = df.iloc[i]['id']
    x = str(post)
    comment_df = pd.read_csv("Comments Corpus.csv"+x)
    comment_data = np.array(comment_df['Comment Score'])
    data = np.concatenate((data, comment_data))
    
fig, ax = plt.subplots(1, 1)
ax.hist(data, bins = 15)
ax.set_xlabel('Score')
ax.set_ylabel('Number of Comments')

plt.show()
```

```python
positive = []
negative = []
neutral = []

p = pd.read_csv("Positive Comments.csv")
n = pd.read_csv("Negative Comments.csv")
d = pd.read_csv("Neutral Comments.csv")

positive = list(p['score'])
negative = list(n['score'])
neutral = d['score']

plt.hist(negative, label='Negative', alpha=0.6, edgecolor='blue', bins = 10)
plt.hist(neutral, label='Neutral', alpha=0.8, edgecolor='yellow', bins = 10)
plt.hist(positive, label='Positive', alpha=1, edgecolor='green', bins = 10)
plt.xlabel('Score')
plt.ylabel('Number of comments')
plt.legend()
 
# Showing the plot using plt.show()
plt.show()
```
