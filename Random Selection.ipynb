```python
import pandas as pd
positive = {
    "post" : [],
    "comment_number" : [],
    "score" :[],
    'text' : []
}
neutral = {
    "post" : [],
    "comment_number" : [],
    "score" :[],
    'text' : []
}
negative = {
    "post" : [],
    "comment_number" : [],
    "score" :[],
    'text' : []
}
```

```python
for i in range(100):
    print(i)
    post = df.iloc[i]['id']
    x = str(post)
    comment_df = pd.read_csv("C:\\Users\\dhruv\\OneDrive\\Desktop\\IITGN\\NLP\\Assignment 1\\Comments\\"+x)
    for j in range(len(comment_df)):
        t = comment_df.iloc[j]['Comment Text']
        s = comment_df.iloc[j]['Comment Score']
        if(comment_df.iloc[j]['label']=='positive'):
            positive["post"].append(post)
            positive["comment_number"].append(j)
            positive['score'].append(s)
            positive['text'].append(t)
        elif(comment_df.iloc[j]['label']=='neutral'):
            neutral["post"].append(post)
            neutral["comment_number"].append(j)
            neutral['text'].append(t)
            neutral['score'].append(s)
        else:
            negative["post"].append(post)
            negative["comment_number"].append(j)
            negative['text'].append(t)
            negative['score'].append(s)
```

```python
positive = pd.DataFrame.from_dict(positive)
negative = pd.DataFrame.from_dict(negative)
neutral = pd.DataFrame.from_dict(neutral)

positive.to_csv("C:\\Users\\dhruv\\OneDrive\\Desktop\\IITGN\\NLP\\Assignment 1\\Positive Comments")
negative.to_csv("C:\\Users\\dhruv\\OneDrive\\Desktop\\IITGN\\NLP\\Assignment 1\\Negative Comments")
neutral.to_csv("C:\\Users\\dhruv\\OneDrive\\Desktop\\IITGN\\NLP\\Assignment 1\\Neutral Comments")
```

```python
p = len(positive['post'])
n = len(negative['post'])
d = len(neutral['post'])
```

```python
import random
```

```python
manual = {'Comment_Text' : [], 
         'Post_Id' : [], 
         'Comment_Number' : [], 
         'label_model' : []}

p_sample = random.sample(range(p), 34)
n_sample = random.sample(range(n), 33)
d_sample = random.sample(range(d), 33)

for i in p_sample:
    post = positive['post'][i]
    c = positive['comment_number'][i]
    manual['Post_Id'].append(post)
    manual['Comment_Number'].append(c)
    manual['label_model'].append('positive')
    manual['Comment_Text'].append(positive['text'][i])

for i in n_sample:
    post = negative['post'][i]
    c = negative['comment_number'][i]
    manual['Post_Id'].append(post)
    manual['Comment_Number'].append(c)
    manual['label_model'].append('negative')
    manual['Comment_Text'].append(negative['text'][i])
    
for i in d_sample:
    post = neutral['post'][i]
    c = neutral['comment_number'][i]
    manual['Post_Id'].append(post)
    manual['Comment_Number'].append(c)
    manual['label_model'].append('neutral')
    manual['Comment_Text'].append(neutral['text'][i])

manual = pd.DataFrame.from_dict(manual)
manual.to_csv("C:\\Users\\dhruv\\OneDrive\\Desktop\\IITGN\\NLP\\Assignment 1\\Manual")
```
